```python
def clean_rest_name(name):
    if isinstance(name,str):
        if re.findall(r'[a-zA-Z0-9 ]',name):
            list_name = re.findall(r'[a-zA-Z0-9 ]',name)
        string_name = "".join(list_name)
        return string_name
    else:
        return name
```

```python
after_droping_duplicates['name'] = after_droping_duplicates['name'].apply(clean_rest_name)
```

```python
def clean_rate(rate):
    if rate == "-" or rate == "NEW":
        return np.nan
    else:
        rate_converted =float(str(rate).split('/')[0])
        return rate_converted
```
